```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# using this library to make a chorlopleth map. if you don't have it you can pip install, but this is an optional library only used for one graph
import plotly.graph_objs as go
```

# Load in Dataset school_scores_and_more
#### SAT FROM HERE BASED OFF OF 2400 not 1600 (but writing category is not included so an additonal 800 points are missing)

```python
school_scores_and_more = pd.read_csv('../data/school_scores_and_more.csv')
school_scores_and_more.head()
```

```python
# see which columns i want to keep or make new df from
school_scores_and_more.info()
```

# Found idea of going based off of wealth 

```python
#based on the df above i made a smaller df with just the family income
family_incomes = school_scores_and_more.iloc[ : , 19:43 ]
family_incomes.head(3)
```

```python
#just want to see if i can get the total of the reading and math to work with
total_by_wealth = family_incomes.loc[: , family_incomes.columns.str.contains('Total')]
total_by_wealth.head(3)
```

```python
# I want to get State and Year in the front using insert method
# DataFrameName.insert(loc, column, value, allow_duplicates = False)
total_by_wealth.insert(0, 'State', school_scores_and_more['State Name'])
total_by_wealth.insert(1, 'Year', school_scores_and_more['Year'])
total_by_wealth
```

```python
total_by_wealth.loc[:, 'Total Reading and Math 20-40k': ]
```

```python
# Lets also get the mean for each row
pd.options.mode.chained_assignment = None   #default='warn'

total_by_wealth['Average of all'] = round(total_by_wealth.loc[:, 'Total Reading and Math 20-40k': ].mean(axis = 1),2)
total_by_wealth
```

## Total score by wealth

```python
#lets also move the <20k

#going to pop it so i can add it in the front without doubles
less_than_20k = total_by_wealth.pop('Total Reading and Math less than 20k')

total_by_wealth.insert(2, 'Total Reading and Math less than 20k', less_than_20k)
total_by_wealth.head()
```

## Total by year and state

```python
#grouped by year and state for the sake of comparison with capita_wealth_by_state
year_and_state = total_by_wealth.groupby(['Year', 'State']).mean()
year_and_state.head()
```

```python
year_and_state.plot(kind = 'hist', x='Average of all', bins = 50, ec = 'Black', figsize = (12, 6));
plt.legend(bbox_to_anchor=(1, 1));
plt.axis([700, 1300, 0, 102]);
```

## Make diff data set with just the state and total scores

```python
total_by_state = total_by_wealth.groupby('State').mean()
# lets drop the year column since it will already avg all the years when doing this
total_by_state.drop('Year', axis = 1, inplace = True)
total_by_state.head(1)
```

```python
total_by_state.plot(y = 'Average of all', kind = 'hist', figsize = (10, 5), bins = 18);
plt.axvline(total_by_state['Average of all'].mean(), color = 'red', label = 'Mean');
plt.legend();
```

```python
#another groupby just by year and see the mean not including state
by_year_wealth_scores = total_by_wealth.groupby('Year').mean()
by_year_wealth_scores.head(3)
```

## Well got the figsize working

```python
# code idea used from geeks for geeks site linked below
# https://www.geeksforgeeks.org/plotting-multiple-bar-charts-using-matplotlib-in-python/

ind = np.arange(2005,2016) # Get range values equal to the year
width = 0.13 # set width size of the bars to be able to space them

plt.figure(figsize = (20,16), dpi = 900)

#set the values to make code for bar a bit cleaner
val1 = by_year_wealth_scores['Total Reading and Math less than 20k']
val2 = by_year_wealth_scores['Total Reading and Math 20-40k']
val3 = by_year_wealth_scores['Total Reading and Math 40-60k']
val4 = by_year_wealth_scores['Total Reading and Math 60-80k']
val5 = by_year_wealth_scores['Total Reading and Math 80-100k']
val6 = by_year_wealth_scores['Total Reading and Math more than 100k']


# lets add the bars one at a time
# also changed the colors because the default were hurting my eyes
bar1 = plt.bar(ind, val1, width, color = '#FC8787')
bar2 = plt.bar(ind + width, val2, width, color = '#FCC987')
bar3 = plt.bar(ind + width*2, val3, width, color = '#C27636')
bar4 = plt.bar(ind + width*3, val4, width, color = '#BBE54A')
bar5 = plt.bar(ind + width*4, val5, width, color = '#4985A2')
bar6 = plt.bar(ind + width*5, val6, width, color = '#78FF80')

# Labels for the graph
plt.xlabel('Year', fontsize = 20)
plt.ylabel('Score', fontsize = 20)
plt.title('Scores Based Off Financial Groups', fontsize = 24)

# Lables for the bars using legend
plt.legend( (bar1, bar2, bar3, bar4, bar5, bar6) ,
            ('Total Reading and Math less than 20k', 'Total Reading and Math 20-40k',
             'Total Reading and Math 40-60k', 'Total Reading and Math 60-80k',
             'Total Reading and Math 80-100k','Total Reading and Math more than 100k'),
              fontsize = 18,
              loc= 2, bbox_to_anchor=(1, 1), ncol= 1);

plt.xticks(ind + width*2.5, list(range(2005,2016)), fontsize = 16, rotation = 20);
plt.yticks(fontsize = 16);
```

### Lets make another df of test takers based on wealth

```python
test_takers = family_incomes.loc[: , family_incomes.columns.str.contains('Test-takers')]
test_takers.head(1)
```

```python
#gotta add state and years now again
test_takers.insert(0, 'State', school_scores_and_more['State Name'])
test_takers.insert(1, 'Year', school_scores_and_more['Year'])
test_takers.head(1)
```

```python
#lets also move the <20k to the front of the df
less_than_20k_takers = test_takers.pop('Family Income Less than 20k Test-takers')
test_takers.insert(2, 'Family Income Less than 20k Test-takers', less_than_20k_takers)

test_takers.head(3)
```

```python
# Mean of all States and graphed based on participation by wealth
test_takers.groupby(['Year']).mean().plot(kind = 'bar', figsize=(18,15), width = .7, fontsize = 18);
plt.legend(loc= 2, bbox_to_anchor=(1, 1), ncol= 1, fontsize = 15);
plt.title('Participation by Wealth Groups', fontsize = 24, fontname= 'Comic sans Ms');
plt.xlabel('Year', fontsize = 20);
plt.ylabel('Number of Participants', fontsize = 20);
```

```python
# This code is taken with minor modifications from https://seaborn.pydata.org/generated/seaborn.heatmap.html

# Establish size of figure.
plt.figure(figsize= (15,10))

# Get correlation of variables.
corr = test_takers.corr()

# Set up mask to be "True" in the upper triangle.
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True

# Plot our correlation heatmap, while masking the upper triangle to be white.
sns.heatmap(corr,
            cmap = 'mako', #cmap is the palette
            annot = True,  #annot is the numbers inside the box
            mask = mask, # applying the mask we made
           linecolor ='white', #lines between the boxes
           linewidths = 0.075) #width of the line between the boxes

plt.xticks(rotation = -80)

plt.show()
```

## Conclusion
> There is a strong correlation between the wealth gaps but little correlation towards years and wealth. The farther Apart the wealth gap is, the less correlation they have; however as you go up each bracket, the correlation between them goes up as well. Almost like a correlation between the correlation!

```python
wealth_percent = pd.read_csv('../data/Income dist.csv')
wealth_percent
```

```python
wealth_percent['Percentage of U.S. households'] = wealth_percent['Percentage of U.S. households'].str.replace('%','').astype(float)
wealth_percent
```

```python
wealth_percent.loc[6, 'Percentage of U.S. households'] = wealth_percent.loc[6:, 'Percentage of U.S. households'].sum()
wealth_percent.loc[6, 'Annual household income in U.S. dollars'] = '100,000 and over'
wealth_percent.drop([7,8], inplace = True)
```

```python
wealth_percent.loc[1,['Percentage of U.S. households']] += 9.1
```

```python
wealth_percent.loc[1,['Annual household income in U.S. dollars']] = '24,999 and Below'
```

```python
wealth_percent.drop(0, inplace = True)
```

```python
tt_mean = test_takers.iloc[:,2:]
tt_mean.head()
```

```python
plt.figure(figsize = (14,14))

colors = ['yellowgreen','red','gold','lightskyblue','lightcoral','grey']

ax1 = plt.subplot(1,2,1);
ax1 = plt.pie(data = wealth_percent,
              x = 'Percentage of U.S. households',
              colors = colors,
              autopct='%1.1f%%');
ax1 = plt.legend(labels = wealth_percent['Annual household income in U.S. dollars'],
                 loc='lower left', bbox_to_anchor=(0.25, -.3),fontsize=12);
ax1 = plt.title('Population Percent by Wealth', fontsize = 15);


ax2 = plt.subplot(1,2,2);
ax2 = plt.pie(data = tt_mean,
              x = tt_mean.mean(),
              colors = colors,
              autopct='%1.1f%%');
ax2 = plt.legend(labels = tt_mean.mean().index,
                 loc='lower right', bbox_to_anchor=(.95, -.3),fontsize=12);
ax2 = plt.title('Test Takers Population by Wealth', fontsize = 15);
```
